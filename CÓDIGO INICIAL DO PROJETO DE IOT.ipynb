```python
//--------------------LCD 16X2---------------------
#include <Wire.h>
#include <LiquidCrystal_I2C.h>

//--------------------LCD 16X2---------------------
#define endereco  0x27
#define colunas   16
#define linhas    2
LiquidCrystal_I2C lcd(endereco, colunas, linhas);

//-----------SENSOR ULTRASSÔNICO E BUZZER----------
#define trigPin 7
#define echoPin 8
#define tempo 10
int frequencia = 0;
int Buzzer = 13;
int duration, distance;

//---------SENSOR DE PRESENÇA E LED AMARELO--------
int pinoledamarelo = 2;
int pinopir = 4;
int acionamento;

//###################################################//
void setup() {
//--------------------LCD 16X2---------------------
  lcd.init();
  lcd.backlight();
  lcd.clear();

//-----------SENSOR ULTRASSÔNICO E BUZZER----------
  pinMode(trigPin, OUTPUT);
  pinMode(echoPin, INPUT);
  pinMode(Buzzer, OUTPUT);

//---------SENSOR DE PRESENÇA E LED AMARELO--------
  pinMode(pinoledamarelo, OUTPUT);
  pinMode(pinopir, INPUT);
}

//###################################################//
void loop() {

  digitalWrite(trigPin, HIGH);
  delayMicroseconds(10);
  digitalWrite(trigPin, LOW);
  duration = pulseIn(echoPin, HIGH);
  distance = (duration / 2) / 29.1;
  acionamento = digitalRead(pinopir);

//-------ACIONAMENTO DO SENSOR DE PRESENÇA E LED AMARELO--------
if (acionamento == LOW) {
    digitalWrite(pinoledamarelo, LOW);
  }

  else {
    digitalWrite(pinoledamarelo, HIGH);
    lcd.setCursor(0,1);
    lcd.print("Movi. Detectado!");
    delay(1000);
    lcd.clear();
}
//--------ACIONAMENTO DO SENSOR ULTRASSÔNICO E BUZZER----------
if (distance >= 20 || distance <= 0) {

    lcd.setCursor(0,0);
    lcd.print("Seguro!");
    digitalWrite(Buzzer, LOW);
 }
  else {

    lcd.clear();
    lcd.setCursor(0,0);
    lcd.print("Aprox. Detectada!");
    lcd.setCursor(0,1);
    lcd.print("Distancia: ");
    lcd.print(distance);
    lcd.setCursor(14,1);
    lcd.print("cm");


  for (frequencia = 150; frequencia < 1800; frequencia += 1) {
      tone(Buzzer, frequencia, tempo);
      delay(3);

 }
  for (frequencia = 1800; frequencia > 150; frequencia -= 1) {
      tone(Buzzer, frequencia, tempo);
      delay(3);
  }
      lcd.clear();
 }
 }
```
